```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("datasets/modern_RAPTOR_by_player.csv")
table = df.pivot_table("poss", index="player_name", columns="season")
df_nw = df[df.season == 2014]
print(table.head())
plt.style.use("seaborn")
```

```python
%matplotlib inline
table.head().plot(kind="bar", rot=45, title="Players vs Poss", ylabel="poss")
```

```python
%matplotlib inline
df_nw.head().plot(x="player_name", y="poss", rot=45, title="PLayers vs Poss")
```
